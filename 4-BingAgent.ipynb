### Setup

1. Create Bing Grounding Connection in project
2. Update the connection to env: BING_CONNECTION_NAME

### Connect to Foundry project

```python
from util import get_project_client
project = get_project_client()
```

```python
import os
from azure.ai.agents.models import BingGroundingTool

bing_connection = project.connections.get(name=os.environ["BING_CONNECTION_NAME"])

# Initialize agent bing tool and add the connection id
bing = BingGroundingTool(connection_id=bing_connection.id)
```

### Create a simple agent

```python
## need to login using the command azd auth login
agent = project.agents.create_agent(
        model=os.getenv("CHAT_MODEL"),
        name="BingAgent",
        instructions="You are helpful AI assistant.",
        tools=bing.definitions,
        tool_resources=bing.resources
    )
print(f"Created agent, agent ID: {agent.id}")
```

```python
from util import sendMessage

sendMessage(project, agent.id, "What is Microsoft's stock price?")
```
